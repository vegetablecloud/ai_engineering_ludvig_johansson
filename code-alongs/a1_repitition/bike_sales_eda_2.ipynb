# Bike Sales EDA part 2

```python
import pandas as pd

df = pd.read_csv("../../data/Sales.csv", index_col=0, parse_dates=True)
df.head()
```

```python
df.sort_index(inplace=True)

df.loc["2013-01-01": "2013-12-01"]
```

```python
# Detta är din Serie
sales_2013 = df.loc['2013']['Country'].value_counts()

# Vi gör om den till en DataFrame
df_sales_2013 = sales_2013.reset_index()

# Låt oss byta namn på kolumnerna till något tydligt
df_sales_2013.columns = ['Country', 'NumberOfRows']

# Titta på resultatet!
df_sales_2013.head()
```
